```python
import pandas as pd

data = '/content/drive/MyDrive/machine learning/Titanic-Dataset.csv' # path dataset
df = pd.read_csv(data) # load dataset
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Age - mean
df['Age'].fillna(value=df['Age'].mean(), inplace=True)

# Cabin - "DECK"
df['Cabin'].fillna(value="DECK", inplace=True)

# Embarked - modus
df['Embarked'].fillna(value=df['Embarked'].mode, inplace=True)
```
